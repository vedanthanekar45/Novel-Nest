Importing libraries

```python
import pandas as pd
import numpy as np
```

Importing the dataset

```python
dataset = pd.read_csv("COVID.csv")
dataset
```

```python
x = dataset['Actual']
y = dataset['Predicted']
```

Calculating the parameters

```python
import sklearn.metrics as sm

accuracy = sm.accuracy_score(x, y)
precision = sm.precision_score(x, y)
recall = sm.recall_score(x, y)
f1_score = sm.f1_score(x, y)
```

Print the parameters

```python
print("Accuracy: ", accuracy)
print("Precision: ", precision)
print("Recall: ", recall)
print("F1 score: ", f1_score)
```

Vizualization

```python
sm.ConfusionMatrixDisplay.from_predictions(x, y)
```
