Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Importing Dataset

```python
dataset = pd.read_csv('Heart.csv')
dataset
```

Handling Duplicates

```python
duplicates = dataset.duplicated().sum()
duplicates

dataset = dataset.drop_duplicates()
# There are no duplicates
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
dataset['AHD'] = le.fit_transform(dataset['AHD'])
dataset
```

One Hot Encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(sparse_output=False), ['ChestPain', 'Thal'])],
    remainder='passthrough'
)

dataset = ct.fit_transform(dataset)
dataset = pd.DataFrame(dataset, columns=ct.get_feature_names_out())
dataset
```

Handling Missing Data

```python
missing = dataset.isnull().sum()
missing

from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy='mean');
dataset = si.fit_transform(dataset)
```

```python
dataset
```

Train Test Split

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(dataset, test_size=0.25, random_state=0)
print("Training data size:", train_data.shape)
print("Testing data size:", test_data.shape)
```
