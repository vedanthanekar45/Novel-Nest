Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Importing the dataset

```python
dataset = pd.read_csv('temperatures.csv')
dataset
```

```python
x = dataset["YEAR"]
```

Making the model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import sklearn.metrics as sm
from sklearn.preprocessing import StandardScaler

lr = LinearRegression()
sc = StandardScaler()
for i in range (1, 13):
    y = dataset.iloc[:, i]
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
    x_train = sc.fit_transform(x_train.values.reshape(-1,1))
    x_test = sc.transform(x_test.values.reshape(-1,1))
    lr.fit(x_train, y_train)
    pred = lr.predict(x_test)
    print("Predicted Values: ", pred)
    mse = sm.mean_squared_error(y_test, pred)
    print("Mean Squared Error: ", mse)
    mae = sm.mean_absolute_error(y_test, pred)
    print("Mean Absolute Error: ", mae)
    r2 = sm.r2_score(y_test, pred)
    print("R2 Score: ", r2)
    print("\n")
    plt.scatter(x_test, y_test, color="blue")
    plt.plot(x_test, pred, color="red", linewidth=2)
    plt.show()
```
